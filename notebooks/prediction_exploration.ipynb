# Prediction Exploration

```python
import pandas as pd
```

```python
features = pd.read_csv("../data/processed/driver_features.csv")
```

```python
features
```

```python
from sklearn.preprocessing import LabelEncoder, StandardScaler
```

```python
le = LabelEncoder()
features["TeamName_enc"] = le.fit_transform(features["TeamName"])
features["GrandPrix_enc"] = le.fit_transform(features["GrandPrix"])
features["Driver_enc"] = le.fit_transform(features["Driver"])
features["Compound_enc"] = le.fit_transform(features["Compound"])
```

```python
features = features.drop(columns=["GrandPrix", "TeamName", "Driver", "Compound"])
```

```python
features
```

```python
scaler = StandardScaler()
X = features.drop(columns=["is_podium"])
X_scaled = scaler.fit_transform(X)
X = pd.DataFrame(X_scaled, columns=X.columns)
X = X.drop(columns=["Position"])

y = features["is_podium"]
```

```python
from sklearn.model_selection import train_test_split
from sklearn.ensemble import RandomForestClassifier
from xgboost import XGBClassifier
from sklearn.metrics import accuracy_score, f1_score, confusion_matrix, classification_report
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42, stratify=y)
```

```python
model = RandomForestClassifier(random_state=42)
model.fit(X_train, y_train)
y_pred = model.predict(X_test)
```

```python
acc = accuracy_score(y_test, y_pred)
f1 = f1_score(y_test, y_pred)

print(f"Accuracy: {acc}")
print(f"F1: {f1}")
```

```python
confusion_matrix(y_test, y_pred)
```

```python
print(classification_report(y_test, y_pred))
```
