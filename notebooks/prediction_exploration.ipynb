```python
times =  session.laps[["Driver", "Sector1SessionTime", "Sector2SessionTime", "Sector3SessionTime", "Team"]].copy()
```

```python
times["Sector1"] = times["Sector1SessionTime"].dt.total_seconds()
times["Sector2"] = times["Sector2SessionTime"].dt.total_seconds()
times["Sector3"] = times["Sector3SessionTime"].dt.total_seconds()
```

```python
times = times.drop(columns=["Team", "Sector1SessionTime", "Sector2SessionTime", "Sector3SessionTime"])
```

```python
times = times.dropna()
times
```

```python
times = times.groupby(["Driver"]).sum()
times
```

```python
times["Total"] = times["Sector1"] + times["Sector2"] +times["Sector3"]
```

```python
times.sort_values(by="Total")
```
